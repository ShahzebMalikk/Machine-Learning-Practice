# Imported all important libraries

```python
import pandas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
import seaborn as sns; sns.set()
```

# Imported Datasets

```python
data = pd.read_csv('Mall_Customers.csv')
```

```python
print(data.head(200))
```

```python
print(data.tail())
```

# 1. Line Plot

Taking Annual Income as X value and Age as why value which helps us figure out that according to what age people 
have how much salary.

```python
plt.plot(data['Annual Income (k$)'].values, data['Age'].values)
```

# 2. Bar Plots

With the help of bar plot we will check who have more spending score male or female.

```python
sns.barplot(x='Gender', y='Spending Score (1-100)', data=data)
```

# 3. Box Plot

we will be using Box plot to check the average of which label ( Male, Female ) have Annual Income in what range.

```python
Male = data[data['Gender'] == 'Male']
Female = data[data['Gender'] == 'Female']
```

```python
feature = 'Annual Income (k$)'

bp_data = [Male[feature], Female[feature]]
plt.boxplot(bp_data, vert=False, labels=['Male', 'Female'])
plt.show()
```

```python
sns.boxplot(x='Annual Income (k$)', y='Gender', data=data)
plt.show()
```

# 4. Violin Plot

```python
sns.violinplot(x='Gender', y='CustomerID', data=data)
```

```python
sns.violinplot(x='Gender', y='Age', data=data)
```

# 5. Correlation

It shows all the plots for the correlation between all features

```python
features = ['CustomerID', 'Gender', 'Age', 'Annual Income (k$)', 'Spending Score (1-100)']         
```

```python
sns.pairplot(data[features])
```
