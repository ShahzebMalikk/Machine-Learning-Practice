# Importing Libraries

```python
import pandas as pd

import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns;  sns.set()
```

# Generating synthetic dataset

```python
def func(X): return X**2

X = np.arange(0, 30)
y = func(X) + np.random.randint(10, size=(30)) * np.random.normal(100)
data = zip(X, y)
```

```python
import pandas
syn_data = pandas.DataFrame(data, columns=['X', 'y'])
print(syn_data.head(31))
```

```python
print(syn_data.info())
```

```python
print(syn_data.describe())
```

# Line Plot

```python
plt.plot(syn_data['X'].values, syn_data['y'].values)
```

```python
sns.lineplot( x = 'X', y = 'y', data=syn_data)
```
